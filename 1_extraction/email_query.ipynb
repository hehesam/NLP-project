```python
from pathlib import Path
import json


def load_emails_by_id(jsonl_path: str | Path):
    """Load a JSONL file into a dict keyed by the 'id' field."""
    jsonl_path = Path(jsonl_path)
    emails = {}

    with jsonl_path.open("r", encoding="utf-8") as f:
        for line_number, line in enumerate(f, start=1):
            line = line.strip()
            if not line:
                continue
            try:
                row = json.loads(line)
            except json.JSONDecodeError as e:
                print(f"Skipping invalid JSON on line {line_number}: {e}")
                continue

            row_id = str(row.get("id", "")).strip()
            if row_id:
                emails[row_id] = row

    return emails


def print_email_by_ids(ids, emails_index):
    """Print sender email, subject, and body_plain for one or multiple ids."""
    if isinstance(ids, (str, int)):
        ids = [ids]

    for raw_id in ids:
        email_id = str(raw_id).strip()
        record = emails_index.get(email_id)

        print("=" * 90)
        print(f"ID: {email_id}")

        if not record:
            print("Status: NOT FOUND")
            continue

        sender_email = record.get("sender_email") or record.get("from") or "N/A"
        subject = record.get("subject", "N/A")
        body_plain = record.get("body_plain", "N/A")

        print(f"Email address: {sender_email}")
        print(f"Subject: {subject}")
        print("Body:")
        print(body_plain)

    print("=" * 90)


JSONL_PATH = Path("../data/raw/filtered_emails.jsonl")
emails_index = load_emails_by_id(JSONL_PATH)


# Example usage:
print_email_by_ids("f84152a0a795e014", emails_index)
# print_email_by_ids(["4e5d730277a0ddab", "d8a3699e03d7335a"], emails_index)
```
